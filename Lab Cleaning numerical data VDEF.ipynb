```python
# 9. Busque valores atÃ­picos en las variables continuas. ( Pista : Â¡hay un buen argumento para hacer eso!). En caso de que encuentres valores atÃ­picos, comenta quÃ© harÃ¡s con ellos.

# Seleccionar solo las columnas continuas (numÃ©ricas)
columnas_continuas = df.select_dtypes(include=['int64', 'float64'])

# Calcular el rango intercuartÃ­lico (IQR) para cada columna continua
Q1 = columnas_continuas.quantile(0.25)
Q3 = columnas_continuas.quantile(0.75)
IQR = Q3 - Q1

# Identificar valores atÃ­picos en las columnas continuas
outliers = ((columnas_continuas < (Q1 - 1.5 * IQR)) | (columnas_continuas > (Q3 + 1.5 * IQR))).any(axis=1)

# Mostrar las filas con valores atÃ­picos en las columnas continuas
print("Filas con valores atÃ­picos en las columnas continuas:")
print(columnas_continuas[outliers])
```
