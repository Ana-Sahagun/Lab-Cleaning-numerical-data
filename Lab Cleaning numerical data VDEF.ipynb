```python
# 9. Busque valores atípicos en las variables continuas. ( Pista : ¡hay un buen argumento para hacer eso!). En caso de que encuentres valores atípicos, comenta qué harás con ellos.

# Seleccionar solo las columnas continuas (numéricas)
columnas_continuas = df.select_dtypes(include=['int64', 'float64'])

# Calcular el rango intercuartílico (IQR) para cada columna continua
Q1 = columnas_continuas.quantile(0.25)
Q3 = columnas_continuas.quantile(0.75)
IQR = Q3 - Q1

# Identificar valores atípicos en las columnas continuas
outliers = ((columnas_continuas < (Q1 - 1.5 * IQR)) | (columnas_continuas > (Q3 + 1.5 * IQR))).any(axis=1)

# Mostrar las filas con valores atípicos en las columnas continuas
print("Filas con valores atípicos en las columnas continuas:")
print(columnas_continuas[outliers])
```
